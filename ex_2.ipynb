```python
import numpy as np
import pandas as pd

def candidate_elimination(training_examples):
    num_attributes = len(training_examples[0]) - 1





    S = [np.array(['0'] * num_attributes)] # Initialize with the most specific hypothesis
    G = [np.array(['?'] * num_attributes)] # Initialize with the most general hypothesis

    # Process each training example
    for i, example in enumerate(training_examples):
        features = np.array(example[:-1])
        label = example[-1]

        if label == 'Yes':

            G = [g for g in G if is_consistent(g, features)]


            for s_idx, s in enumerate(S):
                if not is_consistent(s, features):
                    S[s_idx] = generalize_hypothesis(s, features)


            S = remove_redundant(S)

        else:

            S = [s for s in S if not is_consistent(s, features)]


            new_G_hypotheses = []
            for g in G:
                if is_consistent(g, features):
                    specializations = specialize_hypothesis(g, features, training_examples)
                    for spec in specializations:

                        if all(is_more_general_than(spec, s) for s in S):
                            new_G_hypotheses.append(spec)
            G = new_G_hypotheses


            G = remove_redundant(G)

        print(f"Iteration {i+1}:")
        print(f"S: {S}")
        print(f"G: {G}")
        print("--------------------------------")

    return S, G

def is_consistent(hypothesis, example_features):
    for i in range(len(hypothesis)):
        if hypothesis[i] != '?' and hypothesis[i] != example_features[i]:
            return False
    return True

def generalize_hypothesis(hypothesis, example_features):
    new_hypothesis = hypothesis.copy()
    for i in range(len(new_hypothesis)):
        if new_hypothesis[i] == '0':
            new_hypothesis[i] = example_features[i]
        elif new_hypothesis[i] != '?' and new_hypothesis[i] != example_features[i]:
            new_hypothesis[i] = '?'
    return new_hypothesis

def specialize_hypothesis(hypothesis, example_features, all_examples):
    specializations = []
    for i in range(len(hypothesis)):
        if hypothesis[i] == '?':
            for value in set([ex[i] for ex in all_examples]): # Get all possible values for this attribute
                if value != example_features[i]:
                    new_g = hypothesis.copy()
                    new_g[i] = value
                    specializations.append(new_g)
        elif hypothesis[i] != '0' and hypothesis[i] != example_features[i]:
            # If it's already specific and not matching, no specialization possible for this part
            pass
    return specializations

def is_more_general_than(h1, h2): # h1 is more general than h2
    for i in range(len(h1)):
        if h1[i] == '0':
            if h2[i] != '0':
                return False
        elif h1[i] == '?':
            # '?' is more general than any specific value or '0'
            continue
        elif h1[i] != h2[i] and h2[i] != '?': # If h1 has a specific value, h2 must have the same or '?'
            return False
    return True

def remove_redundant(hypotheses_list):
    unique_hypotheses = []
    for h in hypotheses_list:
        is_redundant = False
        for unique_h in unique_hypotheses:
            if all(h == unique_h):
                is_redundant = True
                break
        if not is_redundant:
            unique_hypotheses.append(h)
    return unique_hypotheses
```

```python
data = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Rainy', 'Cold', 'High', 'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', 'Yes']
]

df = pd.DataFrame(data, columns=['Sky', 'AirTemp', 'Humidity', 'Wind', 'Water', 'Forecast', 'EnjoySport'])
display(df)
```

```python
S_final, G_final = candidate_elimination(data)

print("Final Specific Boundary (S):")
for s in S_final:
    print(s)

print("\nFinal General Boundary (G):")
for g in G_final:
    print(g)
```
